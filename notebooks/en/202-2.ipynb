## Example: Winter Olympics, Sochi 2014

```python
import pandas as pd
import seaborn as sns
data = pd.read_table('../data/athletes.tab', skiprows=[1])
data['birthdate'] = pd.to_datetime(data['birthdate'])
data['bmi'] = data['weight'] / (data['height']) ** 2
data['age_group'] = pd.cut(data['age'], [0, 20, 30, 40, 100], labels=["<20", "20-30", "30-40", "40+"])
```

##### Answer 2-2-1

```python
filtered_data = data[(data['sport'] == 'Ice Hockey') | (data['sport'] == 'Ski Jumping')]
sns.displot(filtered_data, x='weight', hue='sport', stat='density');
```
